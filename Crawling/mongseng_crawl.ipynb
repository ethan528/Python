```python
import pandas as pd
from tqdm import tqdm
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
```

```python
# Chrome 드라이버 경로 지정
driver_path = 'Crawling\chromedriver.exe'

# 크롬 옵션 설정
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36")

# 크롤링할 URL 리스트
urls = [f"https://www.mongseng.io/toktok/{i}" for i in range(42016, 41900, -1)]

# 결과를 저장할 리스트 생성
main_test_list = []
comments_list = []

try:
    # Chrome 드라이버 초기화
    driver = webdriver.Chrome(service=Service(driver_path), options=chrome_options)

    for url in tqdm(urls):
        # 웹 페이지 접속
        driver.get(url)
        
        # 페이지 소스코드 가져오기
        html = driver.page_source

        # BeautifulSoup 객체 생성
        soup = BeautifulSoup(html, 'html.parser')
        
        main_text = soup.select("#__next > div > div.toktok-details_container___FDIt > p")
        main_text = ''.join([elem.text.replace('\n', '') for elem in main_text])  # '\n'을 빈 문자열로 대체
        
        main_test_list.append(main_text)

        # 댓글 텍스트를 저장할 리스트
        comments = []

        # 댓글이 위치한 부모 요소 선택
        parent_selector = '.comment_content__Fhvi7'

        # 부모 요소에서 모든 스팬 요소 선택 및 댓글 추출
        for parent_element in soup.select(parent_selector):
            # 두 번째 스팬 요소의 텍스트 가져오기
            comment_text = parent_element.select_one('span:nth-child(2)').text.strip().replace('\n', '')
            comments.append(comment_text)
            
        comments_list.append(comments)
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
    pass

# 드라이버 종료
driver.quit()
```
